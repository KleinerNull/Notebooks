```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
points = np.random.rand(100, 3)
df = pd.DataFrame(points, columns=['x', 'y', 'z'])
df.head()
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked=True, figsize=(20,10))
```

```python
df.mean().plot()
df.median().plot()
```

```python
df.y.plot(kind='hist')
```

```python
df_sum = df.x + df.y + df.z
df['sum'] = df_sum
df[df['sum'] > 1]
```
