```python
# IMPORTANT

# Currently xmldataset has problems with the xmlns='' attribute in the GetFulfillmentOrderResponse node.
# So I replace it manually

annoying_attribute = search(r'xmlns="(.+)"', xml).group() # crazy regex skills!
if annoying_attribute:
    xml = xml.replace(annoying_attribute, '')
```

```python
profile = """
GetFulfillmentOrderResponse
    GetFulfillmentOrderResult        
            FulfillmentOrder
                SellerFulfillmentOrderId = dataset:fulfillment_order
                DestinationAddress
                    CountryCode = dataset:fulfillment_order
                    PostalCode = dataset:fulfillment_order
                    Name = dataset:fulfillment_order
                    Line1 = dataset:fulfillment_order
                    Line2 = dataset:fulfillment_order
                FulfillmentOrderStatus = dataset:fulfillment_order
                MarketplaceId = dataset:fulfillment_order
            FulfillmentShipment
                member
                    FulfillmentShipmentStatus = dataset:fulfillment_shipment
                    AmazonShipmentId = dataset:fulfillment_shipment
                    ShippingDateTime = dataset:fulfillment_shipment
                    FulfillmentCenterId = dataset:fulfillment_shipment
                    FulfillmentShipmentItem
                        member
                            SellerSKU = dataset:fulfillment_shipment
                            SellerFulfillmentOrderItemId = dataset:fulfillment_shipment
                            Quantity = dataset:fulfillment_shipment
                            PackageNumber = dataset:fulfillment_shipment"""
```

```python
result = parse_using_profile(xml, profile)
result
```

```python
# Looks good, but I want to flatten it more. 
# In the end, the informations of fulfillment_order should be merged with the fulfillment_shipment.

order_info, *_ = result['fulfillment_order']
# order_info is unique, but stored in a list, so I unpack the first and only element.
# order_info = result['fulfillment_order'][0] for py2, but I hate using indecies :P
shipments = result['fulfillment_shipment']

for shipment in shipments:
    shipment.update(order_info)
    
shipments
```
