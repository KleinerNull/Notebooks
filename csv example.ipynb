# Creating the dummy data and save it for demonstartion as csv file.

```python
types = 'invoice,order,return'.split(',')
currencies = 'USD,GBP,EUR'.split(',')
```

```python
from radar import random_datetime
# a lib to generate random datetimes, install it with 'pip install radar'
```

```python
data = [{'ComponentType': choice(types), 
         'Date': random_datetime(start='2016-01-01T00:00:00', stop='2016-12-31T23:59:59'),
         'Currency': choice(currencies),
         'CurrencyAmount': randint(1, 1000)} for _ in range(10**4)]
```

```python
# load this data into a pandas dataframe and save it as data.csv
dummy = pd.DataFrame(data)
```

```python
dummy.head()
```

```python
dummy.to_csv('data.csv', index=False)
```

```python
import numpy as np

x_to_USD = {'EUR': 1.12785,
            'GBP': 1.4353}

def power_filter(ComponentType, Currency, CurrencyAmount):
    if ComponentType == 'invoice':
        return CurrencyAmount * x_to_USD.get(Currency, 1) # default case is USD, so simply 1
    if ComponentType == 'return':
        return -CurrencyAmount * x_to_USD.get(Currency, 1) # returns should be losses
    else:
        return 0
```

```python
df['SalesInUSD'] = np.vectorize(power_filter)(df.ComponentType, df.Currency, df.CurrencyAmount)
```
