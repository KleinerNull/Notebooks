```python
import requests as r
```

```python
req = r.get('http://real-chart.finance.yahoo.com/table.csv?s=MSFT&a=07&b=9&c=1996&d=02&e=13&f=2016&g=d&ignore=.csv')

## Just to simulate the csv data
with open('sample_data.csv', 'w') as infile:
    infile.write(req.content)
```

```python
%matplotlib inline 
# the note book shows the plots now inline the document
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('sample_data.csv')
df.head() # shows the first 5 rows
```

```python
df.index # see date is already converted to datetime
```

```python
df[['Open', 'Close']].plot(figsize=(20, 10))
```

```python
df['daily_diff'] = df.High - df.Low
df[['High', 'Low', 'daily_diff']].plot(figsize=(20,10))
```
